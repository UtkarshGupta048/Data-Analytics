```python
mean_val = df['Amount'].mean()
median_val = df['Amount'].median()
mode_val = df['Amount'].mode()[0]
print(mean_val)
print(median_val)
print(mode_val)
```

```python
sns.countplot(x='Gender', data=df , hue='Marital_Status')
```

```python
sm=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(data=df, x="Age Group", hue='Gender')


for bars in ax.containers:
  ax.bar_label(bars)
plt.tight_layout()
plt.show()
```

```python
df.head(2)
```

```python
total_sales=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Age Group' , ascending=True)
# total_sales

# Disable scientific notation
pd.set_option('display.float_format', '{:,.0f}'.format)

# Now display your DataFrame
print(total_sales)
```

```python
ay=sns.barplot(data=total_sales, x='Age Group', y='Amount')
# pd.set_option('display.float_format','{:,.0f}'.format)
for bars in ay.containers:
  ay.bar_label(bars, labels=[f'{v:.0f}' for v in bars.datavalues])
```

```python
ts=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ts
```

```python
sns.barplot(data=ts, x='Age Group' , y='Amount')

plt.show()
```

```python
total_order=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).reset_index(drop=True)
top_ten_states=total_order.head(10)
```

```python
top_ten_states
```

```python
palette= sns.color_palette('husl', len(top_ten_states))
plt.figure(figsize=(8,6))
sns.barplot(data=top_ten_states, x='State', y='Orders', palette=palette)
plt.xticks(rotation=90)
plt.title("Top Ten Sales States")
plt.tight_layout()
plt.show()
```

```python
palette=sns.color_palette('deep')
sns.barplot(data=top_ten_states, y='State', x='Orders',palette=palette,hue='State')
# plt.xticks(rotation=90)
plt.title("Top Ten Sales States")
plt.show()
```

```python
tot_order_by_occupation=df.groupby(['Occupation'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).reset_index(drop=True)
tot_order_by_occupation
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=tot_order_by_occupation,x='Occupation',y='Orders',palette=palette)
plt.xticks(rotation=90)
plt.title('Total Order on the Basis of Occupation')
plt.tight_layout()
plt.show()
```
